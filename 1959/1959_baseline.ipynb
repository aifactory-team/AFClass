```python
# 데이터를 다운로드 받습니다.

!wget https://raw.githubusercontent.com/aifactory-team/AFClass/master/1959/1959_x_train.csv
!wget https://raw.githubusercontent.com/aifactory-team/AFClass/master/1959/1959_y_train.csv
!wget https://raw.githubusercontent.com/aifactory-team/AFClass/master/1959/1959_x_test.csv
```

```python
# 모델을 학습합니다.

import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

x_train = np.loadtxt('1959_x_train.csv')
y_train = np.loadtxt('1959_y_train.csv')
y_train = to_categorical(y_train)
x_test = np.loadtxt('1959_x_test.csv')

# Build the model

model = Sequential()

model.add(Dense(10, input_shape=(4,), activation='relu', name='fc1'))
model.add(Dense(10, activation='relu', name='fc2'))
model.add(Dense(3, activation='softmax', name='output'))

# Adam optimizer with learning rate of 0.001
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print(model.summary())

# Train the model
model.fit(x_train, y_train, verbose=2, batch_size=5, epochs=200)
```

```python
# 학습한 모델로 시험셋에 대해서 추론하여, 그 결과를 y_pred.csv로 저장합니다.

y_pred = model.predict(x_test)
y_pred = y_pred.argmax(axis=-1)
np.savetxt('y_pred.csv', y_pred, fmt='%d')

# 코랩에서 저장된 y_pred.csv로 태스크에 제출합니다.
```
