알파벳 수화 이미지 데이터셋 불러옵니다.

```python
import numpy as np

dataset = np.load('sing_mnist.npz')
x_train, y_train = dataset['x_train'], dataset['y_train']
x_test, y_test = dataset['x_test'], dataset['y_test']
```

데이터셋 형태 살펴봅니다. 크게 훈련셋과 시험셋으로 구분됩니다.

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

시험셋의 샘플 데이터 10개를 가시화 시켜봅니다.

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=10, figsize=(10, 13), 
                         subplot_kw={'xticks': [], 'yticks': []})

for i, ax in enumerate(axes.flat):
    ax.imshow(x_test[i], cmap = 'gray')
    ax.set_title(str(i) + ': ?')
plt.show()
```

알파벳 수화에 대해서 기존에 알고 계신 분은 각 손 모양 이미지가 의미하는 알파벳을 기록하여 파일로 저장합니다.

```python
import numpy as np

y_pred = ['a', 'a', 'a', 'a', 'a', 'a', 'a', 'a', 'a', 'a']

np.savetxt('y_pred.csv', y_pred, fmt='%c')
```

알파벳 수화를 모르시는 경우에는 시험셋이 아닌 훈련셋을 보고 학습을 해보겠습니다. 훈련셋의 샘플 100개를 손 이미지와 라벨을 함께 표출합니다.


```python
fig, axes = plt.subplots(nrows=10, ncols=10, figsize=(10, 13), 
                         subplot_kw={'xticks': [], 'yticks': []})

alphabat = 'abcdefghijklmnopqrstuvwxyz'

for i, ax in enumerate(axes.flat):
    ax.imshow(x_train[i], cmap = 'gray')
    title = alphabat[y_train[i]]
    ax.set_title(title)
plt.show()
```

훈련셋을 충분히 보고 숙지하거나 시험셋의 샘플과 유사한 것을 찾거나하여 시험셋의 라벨을 기록하여 파일에 저장합니다.

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=10, figsize=(10, 13), 
                         subplot_kw={'xticks': [], 'yticks': []})

for i, ax in enumerate(axes.flat):
    ax.imshow(x_test[i], cmap = 'gray')
    ax.set_title(str(i) + ': ?')
plt.show()
```
